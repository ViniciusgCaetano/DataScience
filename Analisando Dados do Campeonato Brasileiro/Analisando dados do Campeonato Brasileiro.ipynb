|Informação| Descrição |

```python
import pandas as pd
```

```python
dados_partida = pd.read_csv('campeonato-brasileiro-full.csv', sep=',')
```

```python
times_e_estados = dados_partida.copy()[['Mandante','Mandante Estado', 'Arena']]
times_e_estados.rename(columns={'Mandante': 'Time',
                        'Mandante Estado': 'Estado'}, inplace=True)
```

```python
limpo = times_e_estados[times_e_estados['Time'] == 'Palmeiras'].drop_duplicates().copy()
limpo.reset_index(inplace=True)
```

```python
serie_duplicadas = times_e_estados[times_e_estados["Time"]=="Palmeiras"].duplicated()
lista = []
test = 0
for i in serie_duplicadas:
    if i:
        var += 1
    else:
        if test > 0:
            lista.append(var)
        test += 1
        var = 1
lista.append(var)

lista

pd.concat([limpo, pd.Series(lista)], axis=1)
#times_e_estados.drop_duplicates(inplace=True)
#times_e_estados.reset_index(drop=True, inplace=True)
#times_e_estados[times_e_estados['Time'] == 'Corinthians']
```
