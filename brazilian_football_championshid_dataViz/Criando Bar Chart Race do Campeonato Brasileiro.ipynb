## Este é um notebook de estudo da bilbioteca Bar Chart Race
Nosso objetivo aqui será pegar um dataset com informações das partidas do Campeonato Brasileiro de um ano arbitrário e plotar um gráfico animado mostrando a situação dos times ao decorrer do campeonato.

```python
#Importando as bilbiotecas que utilizaremos
import numpy as np
import pandas as pd
import bar_chart_race as bcr
```

```python
#Carregando o dataset original
df = pd.read_csv('campeonato-brasileiro-full.csv')
df
```

```python
'''Aqui iremos criar uma coluna auxiliar com os anos em que aconteceram cada partida, para que
possamos filtrar as temporadas baseadas na temporada em que ocorreram'''
df['Ano'] = pd.DatetimeIndex(df['Data']).year
```

```python
'''Aqui vamos filtrar a tabela para termos apenas os registros da temporada de 2019
sinta-se livre para trocar "2019" por outro ano de sua preferência.'''
df_filtrado = df[df['Ano'] == 2019].reset_index(drop=True)
```

```python
df_filtrado
```

```python
'''Aqui vamos criar as nossas colunas e as linhas do dataframe que iremos utilizar
no nosso gráfico, esta é uma tabela onde as colunas representam os times, e as linhas representam
quantos pontos cada time possui em uma determinada rodada do campeonato'''
times = df_filtrado.Mandante.unique()
rodadas = pd.Index(range(0, len(df_filtrado['Rodada'].unique()) + 1))
```

```python
tabela_por_rodada = pd.DataFrame(index=rodadas,columns=[*times])
tabela_por_rodada
```

```python
'''
Aqui realizaremos o preenchimento das tabelas, iniciaremos preenchendo a primeira linha com zeros.
Após isso, vamos realizar um processo bem simples:
Para cada partida da temporada:
- Caso algum time vença, adicionaremos + 3 na célula[time, rodada] e + 0 na célula[time, rodada]
do time perdedor
- Caso haja empate, adicionaremos + 1 na célulla[time, rodada] dos dois times.
'''
tabela_por_rodada
tabela_por_rodada.iloc[0] = np.zeros(len(tabela_por_rodada.iloc[0])
    , dtype=int)
for num in range(0, len(df_filtrado)):
    partida = df_filtrado.iloc[num]
    if partida.Vencedor != '-':
        tabela_por_rodada.at[partida.Rodada, partida.Vencedor] = tabela_por_rodada.at[partida.Rodada - 1, partida.Vencedor] + 3
        
        if partida.Vencedor != partida.Mandante:
            tabela_por_rodada.at[partida.Rodada, partida.Mandante] = 0 + tabela_por_rodada.at[partida.Rodada - 1, partida.Mandante]
            
        else:
            tabela_por_rodada.at[partida.Rodada, partida.Visitante] = 0 + tabela_por_rodada.at[partida.Rodada - 1, partida.Visitante]    
            
    else:
        tabela_por_rodada.at[partida.Rodada, partida.Mandante] = 1 + tabela_por_rodada.at[partida.Rodada - 1, partida.Mandante]
        tabela_por_rodada.at[partida.Rodada, partida.Visitante] = 1 + tabela_por_rodada.at[partida.Rodada - 1, partida.Visitante]
                     
```

```python
'''
Pronto, temos a tabela que mostra a evolução da pontuação de cada time ao decorrer do campeonato,
o que já nos deixa prontos para criar nosso gráfico animado!
'''
tabela_por_rodada
tabela_por_rodada = tabela_por_rodada.astype('int64')
```

```python
#Vamos 
bcr.bar_chart_race(tabela_por_rodada,
                   n_bars=6,
                   steps_per_period=40,
                   period_length=1500)
```
